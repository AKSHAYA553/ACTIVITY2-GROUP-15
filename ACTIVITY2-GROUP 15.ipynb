```python
import pandas as pd
data=pd.read_csv("diabetes_model.csv")
```

# 1.Create a count plot of outcome

```python
import matplotlib.pyplot as plt
import seaborn as sns
ax=sns.countplot(x='Outcome',data=data)
plt.show()
```

# 2.Analyze the effect of age on outcome using boxplot

```python
sns.boxplot(x = 'Outcome', y = 'Age',  data = data)
plt.show()
```

# 3.Check for the chances of getting diabetes if it is common in family,using the Diabetes Pedigreee Function.

```python
sns.boxplot(x = 'Outcome', y = 'DiabetesPedigreeFunction',  data = data)
plt.show()
```

# 4.Check the realationship with Insulin and Glucose

```python
sns.scatterplot(x='Insulin',y='Glucose',hue='Outcome',data=data)
plt.show()
```

# 5.Create heatmap on Correlation of features in the data

```python
sns.heatmap(data.corr(),annot=True)
plt.xticks(rotation=80)
```
